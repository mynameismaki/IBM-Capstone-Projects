## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(data=df, y='ConvertedComp')
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
import numpy as np
q1 = np.nanpercentile(df['ConvertedComp'], 25)
q3 = np.nanpercentile(df['ConvertedComp'], 75)
iqr = q3-q1
iqr
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bound = q3 + (1.5 * iqr)
lower_bound = q1 - (1.5 * iqr)

print(upper_bound, lower_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df[(df['ConvertedComp'] > upper_bound) | (df['ConvertedComp'] < lower_bound)]
print(len(outliers))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_new = df.copy()
for ind in outliers.index:
    df_new.drop(ind, inplace=True)
    
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.
