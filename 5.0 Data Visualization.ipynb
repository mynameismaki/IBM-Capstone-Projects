## Demo: How to work with database


Download database file.


```python
import requests

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite'
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Specify the local file path where you want to save the downloaded file
    local_filename = 'm4_survey_data.sqlite'

    with open(local_filename, 'wb') as file:
        file.write(response.content)

    print(f"File '{local_filename}' has been successfully downloaded.")
else:
    print(f"Failed to download the file. Status code: {response.status_code}")
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt

df_new = pd.read_sql_query("Select * from master", conn)
df_new.drop(columns=['index'], axis=1, inplace=True, errors='ignore')

sns.histplot(data=df_new, x='ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

sns.boxplot(data=df_new, y='Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

sns.scatterplot(data=df_new, x='Age', y='WorkWeekHrs', )
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import numpy as np

fig, ax = plt.subplots(figsize=(20, 10))
minsize = df_new['Age'].min()
maxsize = df_new['Age'].max()
sns.scatterplot(data=df_new, y="WorkWeekHrs", x="CodeRevHrs", size="Age", sizes=(minsize*15, maxsize*50), ax=ax)

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query = """Select DatabaseDesireNextYear, count(Respondent) as count_r from DatabaseDesireNextYear 
            group by DatabaseDesireNextYear 
            order by count(Respondent) desc limit 5"""
df_databases = pd.read_sql_query(query, conn)

df_databases

fig, ax = plt.subplots()
ax.pie(df_databases['count_r'], labels=df_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', explode=(0.1, 0, 0, 0, 0))
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
df_barplot = df_new.copy()
df_barplot = df_barplot[(df_barplot['Age'] >= 30) & (df_barplot['Age'] <= 35)][['Age', 'WorkWeekHrs', 'CodeRevHrs']]
df_barplot.reset_index(inplace=True)
df_barplot.drop(columns=['index'], inplace=True)
df_barplot = df_barplot.groupby('Age').agg({'WorkWeekHrs' : 'median', 'CodeRevHrs' : 'median'})

df_barplot.plot(kind='bar', stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

df_line = df_new.copy()
df_line = df_line[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)][['Age', 'ConvertedComp']]
df_line = df_line.groupby('Age').agg({'ConvertedComp' : 'median'})

sns.lineplot(data=df_line, x='Age', y='ConvertedComp')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

sns.countplot(data=df_new, y='MainBranch', orient='h')
```
